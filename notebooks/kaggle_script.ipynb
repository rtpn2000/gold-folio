```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\rtpn2\\Projects\\gold-folio\\data\\raw\\GoldPrices _(2013-2023).csv')
```

```python
# Clean headers
df.columns = df.columns.str.strip()

# Inspect raw values first
print(df[["Date", "Price"]].head(10))

# Date
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y", errors="coerce")

# Clean Price very aggressively
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .str.strip()
)

print(df["Price"].head(10).tolist())

# Convert to numeric
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# See which rows failed
print(df[df["Price"].isna()][["Date", "Price"]].head(20))

# Drop bad rows
df = df.dropna(subset=["Date", "Price"]).copy()

# Convert ounce -> gram
df["price_per_gram_usd"] = df["Price"] / 31.1035

# Final frame
df = (
    df[["Date", "price_per_gram_usd"]]
    .rename(columns={"Date": "date"})
    .sort_values("date")
    .drop_duplicates(subset=["date"])
)

print(df.head())
print(df.tail())
```
